```python
import pandas as pd
import matplotlib.pyplot as plt

#Read in all files and get their respective row count
Ages = pd.read_csv('Ages.csv') 
AirQuality = pd.read_csv('AirQuality.csv') 
Asthma = pd.read_csv('Asthma_Clean.csv') 
Bingedrinking = pd.read_csv('Bingedrinking.csv') 
CKD = pd.read_csv('CKD.csv') 
COPD = pd.read_csv('COPD_Clean.csv') 
County_Transportation = pd.read_csv('County_Transport_Data.csv') 
Diabetes = pd.read_csv('Diabetes.csv') 
Fire_Risk = pd.read_csv('Fire_Risk_Cleaned.csv') 
Gender = pd.read_csv('Gender_Clean.csv') 
HeartDisease = pd.read_csv('HeartDisease.csv') 
HighBloodPressure = pd.read_csv('HighBloodPressure.csv') 
HighCholesterol = pd.read_csv('HighCholesterol.csv') 
MentalHealthPoor = pd.read_csv('MentalHealthPoor.csv') 
NoLeisure = pd.read_csv('NoLeisure.csv') 
Number_of_Allergen_Plants = pd.read_csv('Number_of_Allergen_Plants_Clean.csv') 
Obesity = pd.read_csv('Obesity.csv') 
Policy_AirQuality = pd.read_csv('Policy_AirQuality.csv') 
Sleepless = pd.read_csv('Sleepless_Clean.csv') 
Smoking = pd.read_csv('Smoking_Clean.csv') 
Smoking_better = pd.read_csv('Smoking_Prevalence_Clean.csv')
Social_Vulnerability_Index = pd.read_csv('Social_Vulnerability_Index.csv') 
Top_Allergen_County = pd.read_csv('Top_Allergen_County_Clean.csv') 
Top_Allergen_State = pd.read_csv('Top_Allergen_State_Clean.csv') 
Urban_rural_code_size = pd.read_csv('Urban_rural_code_size.csv') 
VisitedDentist = pd.read_csv('VisitedDentist.csv') 
LungDisease = pd.read_csv('LungDisease_Clean.csv') 
name_me = pd.read_csv('state_and_county_fips_master.csv')
```

```python

# Merge the large count DataFrames based Fips


#merged1 = pd.merge(Ages, AirQuality, left_on='CountyFips', right_on='CountyFips', how='inner')
merged2= pd.merge(Ages, County_Transportation, left_on='CountyFips', right_on='CountyFips', how='inner')
merged3= pd.merge(merged2, Fire_Risk, left_on='CountyFips', right_on='CountyFips', how='inner')
merged4= pd.merge(merged3, Gender, left_on='CountyFips', right_on='CountyFips', how='inner')
merged5= pd.merge(merged4, Number_of_Allergen_Plants, left_on='CountyFips', right_on='CountyFips', how='inner')
merged6= pd.merge(merged5, Policy_AirQuality, left_on='CountyFips', right_on='CountyFips', how='inner')
merged7= pd.merge(merged6, Social_Vulnerability_Index, left_on='CountyFips', right_on='CountyFips', how='inner')
merged8= pd.merge(merged7, Top_Allergen_County, left_on='CountyFips', right_on='CountyFips', how='inner')
merged9= pd.merge(merged8, Top_Allergen_State, left_on='CountyFips', right_on='CountyFips', how='inner')
merged10= pd.merge(merged9, Urban_rural_code_size, left_on='CountyFips', right_on='CountyFips', how='inner')
merged11 = pd.merge(merged10, name_me, left_on='CountyFips', right_on='fips', how='inner')
merged12 = pd.merge(merged11, Smoking_better, left_on='CountyFips', right_on='CountyFips', how='inner')

merged25 = pd.merge(merged12, LungDisease, left_on='CountyFips', right_on='CountyFips', how='inner')

#counts after merge check
count_after_merge = len(merged25)
print(count_after_merge)
```
